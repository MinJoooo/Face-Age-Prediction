```python
import numpy as np 
import pandas as pd
import os
from pathlib import Path

import matplotlib.pyplot as plt
```

```python
path = Path('/content/gdrive/My Drive/train/')
filepaths = list(path.glob(r'**/*.jpg'))
filepaths[:5]
```

```python
def getLastFolderName(x): 
    pathwithoutFilename = os.path.split(x)[0] 
    lastFolderName = pathwithoutFilename.split("/")[-1] 
    return lastFolderName

targets = list(map(lambda x: getLastFolderName(x),filepaths)) 
targets[:5]
```

```python
column_filepath = pd.Series(filepaths, name='face_image_path').astype(str)
column_fishType = pd.Series(targets, name='age_type').astype(str)

train = pd.merge(column_filepath, column_fishType, right_index = True, left_index = True)
train= train.sample(frac = 1).reset_index(drop=True) # all shuffle 
train.head()
```

```python
f,a = plt.subplots(nrows=4, ncols=4,figsize=(15, 10),
                        subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(a.flat):
    i = i + 18
    ax.imshow(plt.imread(train.face_image_path[i]))
    ax.set_title(train.age_type[i])
    
plt.tight_layout()
plt.show()
```
